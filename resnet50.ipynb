```
from keras.models import Model
from keras.layers.core import Layer
from keras.layers import Input, Conv2D, BatchNormalization, Activation, MaxPooling2D, ZeroPadding2D, AveragePooling2D, Dense, Add, Flatten
```

```
IMAGE_H, IMAGE_W = 224, 224
```

```
class PoolHelper(Layer):
    
    def __init__(self, **kwargs):
        super(PoolHelper, self).__init__(**kwargs)
    
    def call(self, x, mask=None):
        return x[:,1:,1:,:]
    
    def get_config(self):
        config = {}
        base_config = super(PoolHelper, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))
```

```
input_image = Input(shape=(IMAGE_H, IMAGE_W, 3), name='data')

x = Conv2D(64, (7,7), strides=(2,2), padding='same', name='conv1')(input_image)
x = BatchNormalization(name='bn_conv1')(x)
x = Activation('relu', name='conv1_relu')(x)
x = ZeroPadding2D(padding=(1,1))(x)
x = PoolHelper()(x)
x = MaxPooling2D((3, 3), strides=(2,2), name='pool1')(x)

skip_x = x

skip_x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res2a_branch1')(skip_x)
skip_x = BatchNormalization(name='bn2a_branch1')(skip_x)

x = Conv2D(64, (1,1), strides=(1,1), use_bias=False, name='res2a_branch2a')(x)
x = BatchNormalization(name='bn2a_branch2a')(x)
x = Activation('relu', name='res2a_branch2a_relu')(x)

x = Conv2D(64, (3,3), strides=(1,1), padding='same', use_bias=False, name='res2a_branch2b')(x)
x = BatchNormalization(name='bn2a_branch2b')(x)
x = Activation('relu', name='res2a_branch2b_relu')(x)

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res2a_branch2c')(x)
x = BatchNormalization(name='bn2a_branch2c')(x)

x = Add(name='res2a')([skip_x, x])
x = Activation('relu', name='res2a_relu')(x)

skip_x = x

x = Conv2D(64, (1,1), strides=(1,1), use_bias=False, name='res2b_branch2a')(x)
x = BatchNormalization(name='bn2b_branch2a')(x)
x = Activation('relu', name='res2b_branch2a_relu')(x)

x = Conv2D(64, (3,3), strides=(1,1), padding='same', use_bias=False, name='res2b_branch2b')(x)
x = BatchNormalization(name='bn2b_branch2b')(x)
x = Activation('relu', name='res2b_branch2b_relu')(x)

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res2b_branch2c')(x)
x = BatchNormalization(name='bn2b_branch2c')(x)

x = Add(name='res2b')([skip_x, x])
x = Activation('relu', name='res2b_relu')(x)

skip_x = x

x = Conv2D(64, (1,1), strides=(1,1), use_bias=False, name='res2c_branch2a')(x)
x = BatchNormalization(name='bn2c_branch2a')(x)
x = Activation('relu', name='res2c_branch2a_relu')(x)

x = Conv2D(64, (3,3), strides=(1,1), padding='same', use_bias=False, name='res2c_branch2b')(x)
x = BatchNormalization(name='bn2c_branch2b')(x)
x = Activation('relu', name='res2c_branch2b_relu')(x)

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res2c_branch2c')(x)
x = BatchNormalization(name='bn2c_branch2c')(x)

x = Add(name='res2c')([skip_x, x])
x = Activation('relu', name='res2c_relu')(x)

skip_x = x

skip_x = Conv2D(512, (1,1), strides=(2,2), use_bias=False, name='res3a_branch1')(skip_x)
skip_x = BatchNormalization(name='bn3a_branch1')(skip_x)

x = Conv2D(128, (1,1), strides=(2,2), use_bias=False, name='res3a_branch2a')(x)
x = BatchNormalization(name='bn3a_branch2a')(x)
x = Activation('relu', name='res3a_branch2a_relu')(x)

x = Conv2D(128, (3,3), strides=(1,1), padding='same', use_bias=False, name='res3a_branch2b')(x)
x = BatchNormalization(name='bn3a_branch2b')(x)
x = Activation('relu', name='res3a_branch2b_relu')(x)

x = Conv2D(512, (1,1), strides=(1,1), use_bias=False, name='res3a_branch2c')(x)
x = BatchNormalization(name='bn3a_branch2c')(x)

x = Add(name='res3a')([skip_x, x])
x = Activation('relu', name='res3a_relu')(x)

skip_x = x

x = Conv2D(128, (1,1), strides=(1,1), use_bias=False, name='res3b_branch2a')(x)
x = BatchNormalization(name='bn3b_branch2a')(x)
x = Activation('relu', name='res3b_branch2a_relu')(x)

x = Conv2D(128, (3,3), strides=(1,1), padding='same', use_bias=False, name='res3b_branch2b')(x)
x = BatchNormalization(name='bn3b_branch2b')(x)
x = Activation('relu', name='res3b_branch2b_relu')(x)

x = Conv2D(512, (1,1), strides=(1,1), use_bias=False, name='res3b_branch2c')(x)
x = BatchNormalization(name='bn3b_branch2c')(x)

x = Add(name='res3b')([skip_x, x])
x = Activation('relu', name='res3b_relu')(x)

skip_x = x

x = Conv2D(128, (1,1), strides=(1,1), use_bias=False, name='res3c_branch2a')(x)
x = BatchNormalization(name='bn3c_branch2a')(x)
x = Activation('relu', name='res3c_branch2a_relu')(x)

x = Conv2D(128, (3,3), strides=(1,1), padding='same', use_bias=False, name='res3c_branch2b')(x)
x = BatchNormalization(name='bn3c_branch2b')(x)
x = Activation('relu', name='res3c_branch2b_relu')(x)

x = Conv2D(512, (1,1), strides=(1,1), use_bias=False, name='res3c_branch2c')(x)
x = BatchNormalization(name='bn3c_branch2c')(x)

x = Add(name='res3c')([skip_x, x])
x = Activation('relu', name='res3c_relu')(x)

skip_x = x

x = Conv2D(128, (1,1), strides=(1,1), use_bias=False, name='res3d_branch2a')(x)
x = BatchNormalization(name='bn3d_branch2a')(x)
x = Activation('relu', name='res3d_branch2a_relu')(x)

x = Conv2D(128, (3,3), strides=(1,1), padding='same', use_bias=False, name='res3d_branch2b')(x)
x = BatchNormalization(name='bn3d_branch2b')(x)
x = Activation('relu', name='res3d_branch2b_relu')(x)

x = Conv2D(512, (1,1), strides=(1,1), use_bias=False, name='res3d_branch2c')(x)
x = BatchNormalization(name='bn3d_branch2c')(x)

x = Add(name='res3d')([skip_x, x])
x = Activation('relu', name='res3d_relu')(x)

skip_x = x

skip_x = Conv2D(1024, (1,1), strides=(2,2), use_bias=False, name='res4a_branch1')(skip_x)
skip_x = BatchNormalization(name='bn4a_branch1')(skip_x)

x = Conv2D(256, (1,1), strides=(2,2), use_bias=False, name='res4a_branch2a')(x)
x = BatchNormalization(name='bn4a_branch2a')(x)
x = Activation('relu', name='res4a_branch2a_relu')(x)

x = Conv2D(256, (3,3), strides=(1,1), padding='same', use_bias=False, name='res4a_branch2b')(x)
x = BatchNormalization(name='bn4a_branch2b')(x)
x = Activation('relu', name='res4a_branch2b_relu')(x)

x = Conv2D(1024, (1,1), strides=(1,1), use_bias=False, name='res4a_branch2c')(x)
x = BatchNormalization(name='bn4a_branch2c')(x)

x = Add(name='res4a')([skip_x, x])
x = Activation('relu', name='res4a_relu')(x)

skip_x = x

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res4b_branch2a')(x)
x = BatchNormalization(name='bn4b_branch2a')(x)
x = Activation('relu', name='res4b_branch2a_relu')(x)

x = Conv2D(256, (3,3), strides=(1,1), padding='same', use_bias=False, name='res4b_branch2b')(x)
x = BatchNormalization(name='bn4b_branch2b')(x)
x = Activation('relu', name='res4b_branch2b_relu')(x)

x = Conv2D(1024, (1,1), strides=(1,1), use_bias=False, name='res4b_branch2c')(x)
x = BatchNormalization(name='bn4b_branch2c')(x)

x = Add(name='res4b')([skip_x, x])
x = Activation('relu', name='res4b_relu')(x)

skip_x = x

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res4c_branch2a')(x)
x = BatchNormalization(name='bn4c_branch2a')(x)
x = Activation('relu', name='res4c_branch2a_relu')(x)

x = Conv2D(256, (3,3), strides=(1,1), padding='same', use_bias=False, name='res4c_branch2b')(x)
x = BatchNormalization(name='bn4c_branch2b')(x)
x = Activation('relu', name='res4c_branch2b_relu')(x)

x = Conv2D(1024, (1,1), strides=(1,1), use_bias=False, name='res4c_branch2c')(x)
x = BatchNormalization(name='bn4c_branch2c')(x)

x = Add(name='res4c')([skip_x, x])
x = Activation('relu', name='res4c_relu')(x)

skip_x = x

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res4d_branch2a')(x)
x = BatchNormalization(name='bn4d_branch2a')(x)
x = Activation('relu', name='res4d_branch2a_relu')(x)

x = Conv2D(256, (3,3), strides=(1,1), padding='same', use_bias=False, name='res4d_branch2b')(x)
x = BatchNormalization(name='bn4d_branch2b')(x)
x = Activation('relu', name='res4d_branch2b_relu')(x)

x = Conv2D(1024, (1,1), strides=(1,1), use_bias=False, name='res4d_branch2c')(x)
x = BatchNormalization(name='bn4d_branch2c')(x)

x = Add(name='res4d')([skip_x, x])
x = Activation('relu', name='res4d_relu')(x)

skip_x = x

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res4e_branch2a')(x)
x = BatchNormalization(name='bn4e_branch2a')(x)
x = Activation('relu', name='res4e_branch2a_relu')(x)

x = Conv2D(256, (3,3), strides=(1,1), padding='same', use_bias=False, name='res4e_branch2b')(x)
x = BatchNormalization(name='bn4e_branch2b')(x)
x = Activation('relu', name='res4e_branch2b_relu')(x)

x = Conv2D(1024, (1,1), strides=(1,1), use_bias=False, name='res4e_branch2c')(x)
x = BatchNormalization(name='bn4e_branch2c')(x)

x = Add(name='res4e')([skip_x, x])
x = Activation('relu', name='res4e_relu')(x)

skip_x = x

x = Conv2D(256, (1,1), strides=(1,1), use_bias=False, name='res4f_branch2a')(x)
x = BatchNormalization(name='bn4f_branch2a')(x)
x = Activation('relu', name='res4f_branch2a_relu')(x)

x = Conv2D(256, (3,3), strides=(1,1), padding='same', use_bias=False, name='res4f_branch2b')(x)
x = BatchNormalization(name='bn4f_branch2b')(x)
x = Activation('relu', name='res4f_branch2b_relu')(x)

x = Conv2D(1024, (1,1), strides=(1,1), use_bias=False, name='res4f_branch2c')(x)
x = BatchNormalization(name='bn4f_branch2c')(x)

x = Add(name='res4f')([skip_x, x])
x = Activation('relu', name='res4f_relu')(x)

skip_x = x

skip_x = Conv2D(2048, (1,1), strides=(2,2), use_bias=False, name='res5a_branch1')(skip_x)
skip_x = BatchNormalization(name='bn5a_branch1')(skip_x)

x = Conv2D(512, (1,1), strides=(2,2), use_bias=False, name='res5a_branch2a')(x)
x = BatchNormalization(name='bn5a_branch2a')(x)
x = Activation('relu', name='res5a_branch2a_relu')(x)

x = Conv2D(512, (3,3), strides=(1,1), padding='same', use_bias=False, name='res5a_branch2b')(x)
x = BatchNormalization(name='bn5a_branch2b')(x)
x = Activation('relu', name='res5a_branch2b_relu')(x)

x = Conv2D(2048, (1,1), strides=(1,1), use_bias=False, name='res5a_branch2c')(x)
x = BatchNormalization(name='bn5a_branch2c')(x)

x = Add(name='res5a')([skip_x, x])
x = Activation('relu', name='res5a_relu')(x)

skip_x = x

x = Conv2D(512, (1,1), strides=(1,1), use_bias=False, name='res5b_branch2a')(x)
x = BatchNormalization(name='bn5b_branch2a')(x)
x = Activation('relu', name='res5b_branch2a_relu')(x)

x = Conv2D(512, (3,3), strides=(1,1), padding='same', use_bias=False, name='res5b_branch2b')(x)
x = BatchNormalization(name='bn5b_branch2b')(x)
x = Activation('relu', name='res5b_branch2b_relu')(x)

x = Conv2D(2048, (1,1), strides=(1,1), use_bias=False, name='res5b_branch2c')(x)
x = BatchNormalization(name='bn5b_branch2c')(x)

x = Add(name='res5b')([skip_x, x])
x = Activation('relu', name='res5b_relu')(x)

skip_x = x

x = Conv2D(512, (1,1), strides=(1,1), use_bias=False, name='res5c_branch2a')(x)
x = BatchNormalization(name='bn5c_branch2a')(x)
x = Activation('relu', name='res5c_branch2a_relu')(x)

x = Conv2D(512, (3,3), strides=(1,1), padding='same', use_bias=False, name='res5c_branch2b')(x)
x = BatchNormalization(name='bn5c_branch2b')(x)
x = Activation('relu', name='res5c_branch2b_relu')(x)

x = Conv2D(2048, (1,1), strides=(1,1), use_bias=False, name='res5c_branch2c')(x)
x = BatchNormalization(name='bn5c_branch2c')(x)

x = Add(name='res5c')([skip_x, x])
x = Activation('relu', name='res5c_relu')(x)

x = AveragePooling2D((7, 7), strides=(1,1), name='pool5')(x)

x = Flatten()(x)

x = Dense(1000, activation='softmax', name='fc1000')(x)

model = Model(input_image, x)
model.summary()
```
